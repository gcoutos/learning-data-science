#### _Exercício 1.1_

Escreva duas variáveis do tipo string e escreva uma expressão lógica que verifica se essas variáveis são iguais.

#### _Exercício 1.2_

Além disso, defina duas variáveis numéricas e calcule o resto da divisão de uma pela outra.

```python
num_1 = 5
num_2 = 16

resto = num_2 % num_1

print('O resto da divisáo entre as variáveis é', resto)
```

#### _Exercício 2_

Crie uma lista de strings que possui elementos repetidos e posteriormente converta essa lista para um conjunto. Confira o tamanho da lista e o tamanho do conjunto, após a conversão.

```python
strings_list = ['python', 'css', 'html', 'c++', 'java', 'css', 'python']
print(strings_list)

# Observa-se que após transformar um elemento list em set
# todos os valores duplicados são removidos e sua ordem é alterada
strings_set = set(strings_list)
print(strings_set)
```
