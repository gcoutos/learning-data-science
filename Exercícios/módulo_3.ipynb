#### _Exercício 1_

Construa uma função que recebe uma lista de tamanho 5 como parâmetro. Essa função deve percorrer cada elemento dessa lista e verificar se o elemento é par ou não. Se ele for par, salvar ele em uma nova lista que desve ser retornada com todos os valores pares encontrados.

```python
def even_numbers(numbers):
    if len(numbers) > 5:
        raise Exception("Desculpe, a lista deve ter no máximo 5 itens!")
    elif len(numbers) < 5:
        raise Exception("Desculpe, a lista deve ter no mínimo 5 itens!")
        
    evens_list = []
    for n in numbers:
        if n % 2 == 0:
            evens_list.append(n)
        else:
            continue

    return print(f"Dessa lista {numbers}, esses números são pares: {evens_list}")


even_numbers([1, 2, 3, 4, 5])
even_numbers([50, 70, 66, 43, 100])
```

#### _Exercício 2_
Escreva um código que chama o módulo _random_. Dentro deste módulo utilize a função que gera números aleatórios inteiros para criar uma função customizada que pede para o usuário digitar a quantidade de sorteios que ele quer e roda sorteando essa quantidade de números. Alguns usuários podem não entender que o input é um número e digitar por exemplo "três", o que geraria um erro no código. Então escreva essa função de maneira a tratar esse erro, enviando uma mensagem que ajude o usuário a resolver o problema.

```python
def number_sorting():
    try:
        times = int(input('Quantas vezes você deseja sortear números de 0 a 100?'))
        
        # Essa parte do código impede que o usuário faça 0 sorteios
        if times == 0:
            raise Exception('Digite um número inteiro diferente de 0!')
        else:
            print('Os números sorteados são:')
            while times != 0:
                print(randint(0, 100))
                times -= 1

    except ValueError:
        print('Ops! Você precisa digitar um número inteiro!')


number_sorting()
```

#### _Exercício 3_

Crie uma classe cliente. Esse cliente deve possuir 2 atributos: nome e saldo e dois métodos: depositar e sacar. Usando a classe, crie um cliente, utilize os métodos criados para depositar 100 reais na conta do cliente, que deve começar com saldo 0 e depois do depósito apresentar saldo 100. Na sequência, faça um saque de 20 reais. Fique atento: o seu código deve impedir que o usuário saque mais do que ele tem em saldo.

##### _Resolução sem input:_

```python
class Cliente:
    def __init__(self, nome):
        self.nome = nome
        self.saldo = 0

    def depositar(self, deposito):
        self.saldo += deposito

        print('Depósito de {:.2f}'.format(self.saldo))
        print('Seu saldo atual é de R${:.2f}!'.format(self.saldo))

    def sacar(self, saque):

        if saque > self.saldo:
            print('\nSaldo insuficiente em conta corrente!')
        else:
            self.saldo -= saque
            print('\nSaque de R${:.2f} realizado com sucesso!'.format(saque))
            print('Seu saldo atual é de R${:.2f}'.format(self.saldo))


cliente_1 = Cliente('Guilherme Couto')
cliente_1.depositar(100)
cliente_1.sacar(20)
```

##### _Resolução com input:_

```python
class Cliente:
    def __init__(self, nome):
        self.nome = nome
        self.saldo = 0

    def depositar(self):

        deposito = float(input('Qual o valor do depósito? R$'))

        self.saldo += float(deposito)
        print('\n\tForam depositados R${:.2f}!'.format(self.saldo))
        print('\tSeu saldo atual é de R${:.2f}!'.format(self.saldo))

    def sacar(self):

        saque = float(input('\nQual o valor do saque? R$'))
        
        while saque > self.saldo:
            print('Saldo insuficiente em conta corrente!')
            saque = float(input('\nQual o valor do saque? R$'))

        if saque <= self.saldo:
            self.saldo -= saque
            print('\n\tSaque de R${:.2f} realizado com sucesso!'.format(saque))
            print('\tSeu saldo atual é de R${:.2f}'.format(self.saldo))


cliente_1 = Cliente('Gui')
cliente_1.depositar()
cliente_1.sacar()
```
